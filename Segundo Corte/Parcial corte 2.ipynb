2.1 Análisis Exploratorio de Datos (EDA)

 Pasos para realizar un EDA:

1. Comprensión de los datos:
    - Revisar el diccionario de datos para entender el significado de cada variable.
    - Identificar el tipo de cada variable (numérica, categórica, etc.).
    - Verificar la dimensionalidad de los datos (número de filas y columnas).
 Herramientas: Pandas (para cargar y explorar los datos), Jupyter Notebook (para visualizar los datos).

2. Limpieza de datos:
    - Identificar valores faltantes, duplicados o inconsistencias en los datos.
    - Decidir cómo tratar los valores faltantes (eliminarlos, imputarlos con la media/mediana/moda, etc.).
    - Eliminar registros duplicados.
    - Tratar las inconsistencias en los datos.
 Herramientas: Pandas (para manipular y limpiar los datos).

3. Análisis univariado:
    - Examinar cada variable individualmente.
    - Para variables numéricas: calcular estadísticas descriptivas (media, desviación estándar, percentiles), crear histogramas, boxplots, etc.
    - Para variables categóricas: crear tablas de frecuencia, diagramas de barras, etc.
 Herramientas: Pandas, Matplotlib, Seaborn.

4. Análisis bivariado:
    - Examinar las relaciones entre dos variables.
    - Para variables numéricas: crear scatter plots, correlogramas, etc.
    - Para variables categóricas: crear tablas de contingencia, gráficos de barras agrupadas, etc.
 Herramientas: Pandas, Matplotlib, Seaborn.

5. Análisis multivariado:
    - Examinar las relaciones entre múltiples variables.
    - Crear gráficos de dispersión en 3D, matrices de correlación, etc.
 Herramientas: Pandas, Matplotlib, Seaborn.

2.2 Preprocesamiento de Datos

 Valores faltantes:
 - Imputación: Reemplazar los valores faltantes con la media, mediana o moda de la columna.
 - Eliminación: Eliminar los registros con valores faltantes. La decisión depende del porcentaje de valores faltantes.
 - Modelos de imputación: Usar algoritmos de machine learning para predecir los valores faltantes.

 Valores atípicos:
 - Eliminación: Eliminar los valores atípicos si son pocos y no representan correctamente los datos.
 - Transformación: Aplicar una transformación logarítmica o box-cox para reducir el impacto de los valores atípicos.
 - Capado: Sustituir los valores atípicos por un límite superior o inferior predefinido.

2.3 Modelos de Clasificación

 Para predecir la aprobación o rechazo de préstamos, se podría usar Regresión Logística, árboles de decisión o máquinas de vectores de soporte (SVM).
 Regresión logística: Modelo lineal que predice la probabilidad de que una instancia pertenezca a una clase. Simple de entender y eficiente.
 Ventajas: Interpretable, eficiente computacionalmente.
 Desventajas: Asumir linealidad, sensible a valores atípicos.

2.4 Desbalanceo de Clases

 El desbalanceo de clases ocurre cuando una clase tiene muchas más muestras que otra en el conjunto de datos. Puede hacer que el modelo esté sesgado hacia la clase mayoritaria.
 SMOTE (Synthetic Minority Over-sampling Technique) genera nuevas muestras sintéticas para la clase minoritaria, balanceando las clases.

2.5 Overfitting

 El overfitting ocurre cuando el modelo aprende demasiado bien los datos de entrenamiento y no generaliza bien a los datos nuevos.
 Detección: Se mide el rendimiento del modelo en un conjunto de datos de prueba independiente.
 Medidas para evitarlo:
 - Regularización: penaliza los coeficientes grandes del modelo.
 - Simplificación del modelo: reducir la complejidad del modelo o usar un modelo menos complejo.
 - Validación cruzada: evaluar el modelo en múltiples particiones de los datos.
 - Aumentar los datos de entrenamiento: más datos, mejor generalización.

```python
# prompt: Usando la tabla de datos, entrena dos modelos de clasificaci´on: Regresi´on Log´ıstica y Arbol de ´
# Decisi´on para predecir el estado del pr´estamo (loan status). Compara los modelos utilizando la
# m´etrica ROC AUC. Para esto, sigue los pasos:
# • Preprocesa los datos (conversi´on de variables categ´oricas, manejo de valores faltantes).
# • Divide los datos en un conjunto de entrenamiento y uno de prueba.
# • Entrena los modelos de Regresi´on Log´ıstica y Arbol de Decisi´on ´ .
# • Hay alguna evidencia de overfitting si/no .
# • Calcula la m´etrica ROC AUC para cada modelo y comp´aralos.
# • Si detectas desbalanceo de clases, aplica SMOTE para equilibrar el dataset antes de entrenar los
# modelos. haz que corra

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Cargar el dataset (asumiendo que ya está cargado en df)
# df = pd.read_csv('loan_approval_dataset.csv')

# Preprocesamiento
# Convertir variables categóricas a numéricas usando Label Encoding
label_encoder = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

# Llenar valores faltantes con la media para las columnas numéricas
numeric_cols = df.select_dtypes(include=['number']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())


# Separar características (X) y variable objetivo (y)
X = df.drop('loan_status', axis=1)
y = df['loan_status']

# Aplicar SMOTE para manejar desbalanceo de clases
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Entrenar el modelo de Regresión Logística
logreg_model = LogisticRegression(random_state=42)
logreg_model.fit(X_train, y_train)

# Entrenar el modelo de Árbol de Decisión
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Evaluar los modelos usando ROC AUC
logreg_roc_auc = roc_auc_score(y_test, logreg_model.predict_proba(X_test)[:, 1])
dt_roc_auc = roc_auc_score(y_test, dt_model.predict_proba(X_test)[:, 1])

print(f"ROC AUC para Regresión Logística: {logreg_roc_auc:.4f}")
print(f"ROC AUC para Árbol de Decisión: {dt_roc_auc:.4f}")

# Comparar los modelos y detectar overfitting (se puede hacer un análisis más profundo)
print("Hay alguna evidencia de overfitting? Se requiere más análisis")
```

```python
# prompt: haz la grafica roc

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# ... (Your existing code) ...

# Predict probabilities for the positive class
logreg_probs = logreg_model.predict_proba(X_test)[:, 1]
dt_probs = dt_model.predict_proba(X_test)[:, 1]

# Compute ROC curve and ROC area for each classifier
fpr_logreg, tpr_logreg, _ = roc_curve(y_test, logreg_probs)
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)

fpr_dt, tpr_dt, _ = roc_curve(y_test, dt_probs)
roc_auc_dt = auc(fpr_dt, tpr_dt)

# Plot ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_logreg, tpr_logreg, color='darkorange', lw=2, label=f'Regresión Logística (AUC = {roc_auc_logreg:.2f})')
plt.plot(fpr_dt, tpr_dt, color='navy', lw=2, label=f'Árbol de Decisión (AUC = {roc_auc_dt:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
```

• ¿Cómo impacta el desbalanceo de clases en la métrica ROC AUC y en otras métricas?

El desbalanceo de clases puede afectar significativamente a las métricas de evaluación de un modelo.

* **ROC AUC:**  ROC AUC, en principio, es menos sensible al desbalanceo de clases que otras métricas, ya que considera la probabilidad de clasificación correcta en lugar de la predicción directa. Si bien puede no ser tan afectada, puede ser mejor usar otras métricas. Sin embargo, no es completamente insensible y, en conjuntos de datos muy desequilibrados, su interpretación puede verse ligeramente afectada, ya que se basa en la clasificación correcta de cada clase en términos de probabilidad y no se concentra en la clase minoritaria.

* **Precisión, exactitud, recall, f1-score:** Estas métricas pueden ser engañosas en presencia de desbalanceo. Si un modelo predice la clase mayoritaria en todos los casos, obtendrá una alta precisión si la clase mayoritaria es la predominante en el dataset. Por otro lado, el recall es mejor para ver la capacidad de identificación de la clase minoritaria. Por tanto se debe observar principalmente el *recall* de la clase minoritaria para ver que tan bien esta identificando las instancias de esta. Para este caso, el *f1-score* es bueno ya que combina precisión y recall.

En resumen, el desbalanceo puede llevar a sobreestimar el rendimiento de un modelo que no es capaz de predecir la clase minoritaria correctamente.

• ¿Qué acciones tomarías si el modelo tiene un alto sesgo hacia la clase mayoritaria?

Si el modelo exhibe un alto sesgo hacia la clase mayoritaria, se deben tomar acciones para corregirlo. Algunas estrategias incluyen:

1. **Técnicas de remuestreo:**
    * **Sobremuestreo (Oversampling):** Aumentar la cantidad de instancias de la clase minoritaria. SMOTE (Synthetic Minority Over-sampling Technique), es una técnica popular para generar instancias sintéticas de la clase minoritaria.
    * **Submuestreo (Undersampling):** Disminuir la cantidad de instancias de la clase mayoritaria.  Tome en cuenta que se pierde información.

2. **Ajustar los pesos de las clases:** En muchos algoritmos, se puede ajustar un parámetro para dar mayor peso a la clase minoritaria durante el entrenamiento. Esto ayuda al modelo a considerar la clase minoritaria con más importancia.

3. **Utilizar métricas apropiadas:** Evitar depender solo de la precisión o exactitud. En su lugar, usar métricas como el AUC, *precision*, *recall* y *f1-score* para obtener una evaluación más completa del modelo en el contexto del desbalanceo. En este caso se debe mirar el rendimiento del modelo en la clase minoritaria (rechazos de préstamo).

4. **Recolectar más datos:** Si es posible, obtener más datos de la clase minoritaria para mejorar el balanceo.

5. **Probar diferentes algoritmos:**  Algunos algoritmos pueden ser más robustos al desbalanceo de clases que otros.

La elección de la estrategia dependerá de las características del conjunto de datos y la naturaleza del problema.  La mejor práctica a menudo es probar varias técnicas y comparar sus resultados.

* **Reducción de dimensionalidad:** Explorar técnicas como PCA para reducir el número de variables y minimizar la redundancia.
* **Simplicar el modelo:** Usar modelos más simples, si el rendimiento es similar, esto también puede ayudar a evitar el sobreajuste.


**Impacto del desbalanceo y SMOTE:**

El desbalanceo de clases impactó negativamente en la precisión del modelo, ya que el modelo podía obtener buena precisión simplemente clasificando la mayoría de los ejemplos como la clase mayoritaria. SMOTE ayudó a mitigar este problema mediante la creación de ejemplos sintéticos de la clase minoritaria, lo que permite al modelo aprender de manera más equitativa. Sin embargo, para evaluar el verdadero impacto se deben comparar los resultados de los modelos con y sin SMOTE usando métricas apropiadas, observando el comportamiento del modelo en la clase minoritaria.


**Detección y manejo del overfitting:**

Para detectar el overfitting, se puede comparar el rendimiento del modelo en los conjuntos de entrenamiento y prueba.  Un gran gap en el rendimiento entre ambos sugiere overfitting.  Otras medidas son la curva de aprendizaje, la validación cruzada y análisis de la complejidad del modelo. Las técnicas para manejar el overfitting incluyen regularización,  simplificar el modelo, y aumentar la cantidad de datos de entrenamiento o mejorar la calidad de estos.
