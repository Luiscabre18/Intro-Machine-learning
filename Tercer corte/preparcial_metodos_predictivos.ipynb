#### 4Ô∏è‚É£ Describe **overfitting** y c√≥mo lo detectar√≠as en la pr√°ctica.

*Respuesta:* <!-- TODO -->


overfitting (sobreajuste) es un fen√≥meno en el aprendizaje autom√°tico donde un modelo aprende los datos de entrenamiento "demasiado bien", capturando ruido y detalles espec√≠ficos del conjunto de entrenamiento que no se generalizan a nuevos datos. Esto resulta en un alto rendimiento en el conjunto de entrenamiento pero un rendimiento significativamente peor en datos no vistos (conjunto de prueba).

Para detectar overfitting en la pr√°ctica:

1.  **Comparaci√≥n de m√©tricas de rendimiento:** La forma m√°s com√∫n es comparar el rendimiento del modelo en el conjunto de entrenamiento y en un conjunto de validaci√≥n o prueba. Si el modelo tiene un rendimiento mucho mejor en el conjunto de entrenamiento (por ejemplo, mayor precisi√≥n, menor error) que en el conjunto de prueba, es un fuerte indicio de overfitting.
2.  **Curvas de aprendizaje:** Graficar el error o la m√©trica de rendimiento en funci√≥n del tama√±o del conjunto de entrenamiento puede ayudar. En el caso de overfitting, la curva de entrenamiento mostrar√° un error bajo que disminuye lentamente o se mantiene bajo, mientras que la curva de validaci√≥n mostrar√° un error m√°s alto que puede aumentar o estancarse.
3.  **Validaci√≥n cruzada:** Utilizar t√©cnicas de validaci√≥n cruzada (como k-fold cross-validation) puede dar una estimaci√≥n m√°s robusta del rendimiento del modelo en datos no vistos y ayudar a identificar si el rendimiento var√≠a significativamente entre los diferentes folds.

#### 9Ô∏è‚É£ ¬øQu√© representa el par√°metro *k* en K‚Äëmeans y qu√© ocurre si es muy grande?

*Respuesta:* <!-- TODO -->


En K-means, el par√°metro *k* representa el n√∫mero de cl√∫steres (grupos) en los que queremos dividir los datos. Es el n√∫mero de centroides que el algoritmo intentar√° encontrar para agrupar los puntos de datos.

Si *k* es muy grande:

1.  **Cl√∫steres peque√±os y dispersos:** Cada cl√∫ster contendr√° muy pocos puntos, posiblemente incluso solo uno. Esto puede no proporcionar informaci√≥n √∫til sobre la estructura general de los datos.
2.  **Sobreajuste a los datos:** Un *k* muy grande puede hacer que el algoritmo se "sobreajuste" a los puntos de datos individuales, creando cl√∫steres que se ajustan perfectamente a los datos de entrenamiento pero que no generalizan bien a nuevos datos o a la verdadera estructura subyacente.
3.  **Dificultad de interpretaci√≥n:** Tener un gran n√∫mero de cl√∫steres puede dificultar la interpretaci√≥n de los resultados y la extracci√≥n de *insights* significativos de los grupos.
4.  **Mayor costo computacional:** Aunque el impacto puede no ser tan dram√°tico como en otros algoritmos, aumentar *k* generalmente incrementa ligeramente el costo computacional del algoritmo, especialmente si el conjunto de datos es grande.
5.  **Puede no converger bien:** En algunos casos, con un *k* excesivamente grande, el algoritmo puede tener dificultades para converger a una soluci√≥n estable, o los centroides pueden oscilar entre puntos.

### 4. Limpieza & Feature Engineering

### 4.1 Cree un revie_datetime usando unixReviewTime y la funcion pd.to_datetime

```python
# prompt: Cree un revie_datetime usando unixReviewTime y la funcion pd.to_datetime

df['review_datetime'] = pd.to_datetime(df['unixReviewTime'], unit='s')
df.head()
```

### 4.2 Cree una columna positive que sea igual a 1 cuando overall sea mayor o igual  a 4.

```python
# prompt: Cree una columna positive que sea igual a 1 cuando overall sea mayor o igual a 4.

df['positive'] = (df['overall'] >= 4).astype(int)
df.head()
```

### 4.3 Cree una columna text que sea la union de summary y review text: text_cols['summary'] + '. ' + text_cols['reviewText']).str.strip()

```python
# prompt:  Cree una columna text que sea la union de summary y review text: text_cols['summary'] + '. ' + text_cols['reviewText']).str.strip()

text_cols = df[['summary', 'reviewText']].fillna('') # fillna handles potential NaN values
df['text'] = (text_cols['summary'] + '. ' + text_cols['reviewText']).str.strip()
df.head()
```

### 4.4 Elimine las columnas que no puede usar en el modleo de clasificacion

```python
# prompt: Elimine las columnas que no puede usar en el modleo de clasificacion

# Drop columns not suitable for classification model
# Columns to drop:
# - reviewerID: Unique identifier, not a predictive feature.
# - asin: Product identifier, not a generalizable feature for review sentiment.
# - reviewerName: Personal identifier, not a predictive feature.
# - helpful: Raw count, might have issues with scale or zero values.
# - overall: This is the source of our target variable 'positive', including it would lead to data leakage.
# - summary: Included in the 'text' column.
# - unixReviewTime: Raw timestamp, less useful than derived date features or just the date itself.
# - reviewTime: String representation of date, less useful than 'review_datetime'.
# - day_diff: Potentially leaked information about when the analysis is run, or less relevant than time-based features derived from 'review_datetime'.
# - review_length: Derived feature, less direct than the raw text for TF-IDF.
# - summary_length: Derived feature, less direct than the raw text for TF-IDF.

columns_to_drop = [
    'reviewerID',
    'asin',
    'reviewerName',
    'helpful',
    'overall',
    'summary',
    'unixReviewTime',
    'reviewTime',
    'day_diff',
    'review_length',
    'summary_length',
    'sentimiento', # This is the string version of the target, 'positive' is the numerical one we'll use.
    'reviewText', # Included in the 'text' column.
    'review_datetime' # We are using the text for classification, date information is not used in this simple model.
]

# Drop the specified columns
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore') # errors='ignore' prevents error if a column is not found

# Display the first few rows and info of the cleaned dataframe
print("DataFrame after dropping irrelevant columns:")
print(df_cleaned.head())
print("\nInformation about the cleaned DataFrame:")
df_cleaned.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        df['text'], df['positive'], test_size=0.3, stratify=df['positive'], random_state=42)

pipe = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=10000)),
    ('clf',  LogisticRegression(max_iter=1000, class_weight='balanced', n_jobs=-1, random_state=42))
])

# pipe.fit(X_train, y_train)
```

### 5. Modelo de Clasificaci√≥n ‚Äì Supervisado (25‚ÄØpts)
  * Ralice una regreison logistica y use la columa positive que creoo como target, entregue una matriz de confusion

```python
# prompt: Ralice una regreison logistica y use la columa positive que creoo como target, entregue una matriz de confusion

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Entrenar el pipeline (que ya incluye la regresi√≥n log√≠stica)
pipe.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = pipe.predict(X_test)

# Calcular la matriz de confusi√≥n
cm = confusion_matrix(y_test, y_pred)

# Visualizar la matriz de confusi√≥n
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negativo (0)', 'Positivo (1)'], yticklabels=['Negativo (0)', 'Positivo (1)'])
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.title('Matriz de Confusi√≥n')
plt.show()

# Opcional: Imprimir el reporte de clasificaci√≥n para m√©tricas adicionales (Precision, Recall, F1-score)
print("\nReporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred))
```

```python
# prompt: Haga un modelo para agrupasr los comentarios: vectorizer = TfidfVectorizer(max_features=10000) X_vec = vectorizer.fit_transform(df['text'])
# y cre una variable que se llame clsuter

vectorizer = TfidfVectorizer(max_features=10000)
X_vec = vectorizer.fit_transform(df['text'])

# Aplicar K-means para agrupar
# Decidimos un n√∫mero de clusters, por ejemplo k=3 o k=5.
# Para este ejemplo, usaremos k=5. Puedes experimentar con diferentes valores.
k = 5
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init para evitar advertencia
cluster = kmeans.fit_predict(X_vec)

# A√±adir las etiquetas de cluster al DataFrame original
df['cluster'] = cluster

# Opcional: Evaluar el score de Silueta para tener una idea de la calidad del clustering
# print(f"\nSilhouette Score: {silhouette_score(X_vec, cluster)}")

# Mostrar las primeras filas con la nueva columna 'cluster'
print("\nDataFrame con la columna 'cluster':")
print(df.head())

# Mostrar la distribuci√≥n de los clusters
print("\nDistribuci√≥n de rese√±as por cluster:")
print(df['cluster'].value_counts())
```

```python
# prompt: Perfil de clusters: Entregue una descriptiva de que contenia cada clster en termino de las otras variables

print("\n--- Perfil de Clusters ---")

# Iterar sobre cada cluster para analizar su perfil
for cluster_id in sorted(df['cluster'].unique()):
    print(f"\n--- Cluster {cluster_id} ---")

    # Filtrar el DataFrame para el cluster actual
    cluster_df = df[df['cluster'] == cluster_id]
    num_reviews = len(cluster_df)

    print(f"Total de rese√±as en este cluster: {num_reviews}")

    # An√°lisis de la variable 'overall' (rating)
    print("\nDistribuci√≥n de Overall (Ratings):")
    print(cluster_df['overall'].value_counts(normalize=True).sort_index()) # Mostrar porcentajes ordenados por rating

    # An√°lisis de la variable 'positive' (sentimiento binario)
    print("\nDistribuci√≥n de Sentimiento (Positivo/Negativo):")
    print(cluster_df['positive'].value_counts(normalize=True)) # Mostrar porcentajes

    # An√°lisis de la variable 'helpful' (votos √∫tiles) - estad√≠sticas descriptivas
    # Aseg√∫rate de que 'helpful' es num√©rico. Si es una cadena, puedes necesitar convertirla.
    # Aqu√≠ asumimos que 'helpful' en el df original ya era num√©rico o se limpi√≥ antes del drop.
    # Si usaste df_cleaned para el clustering, podr√≠as necesitar re-unir las columnas relevantes.
    # Si el clustering se hizo sobre df['text'], entonces podemos usar el df original.
    if 'helpful' in cluster_df.columns:
        print("\nEstad√≠sticas descriptivas de Votos √ötiles ('helpful'):")
        print(cluster_df['helpful'].describe())
    else:
        print("\nLa columna 'helpful' no est√° disponible en el DataFrame del cluster.")


    # An√°lisis de la variable 'review_length'
    if 'review_length' in cluster_df.columns:
        print("\nEstad√≠sticas descriptivas de la Longitud de la Rese√±a:")
        print(cluster_df['review_length'].describe())
    else:
         print("\nLa columna 'review_length' no est√° disponible en el DataFrame del cluster.")


    # An√°lisis de las palabras m√°s frecuentes en el texto del cluster
    # Puedes tomar una muestra si el cluster es muy grande
    sample_text = ' '.join(cluster_df['text'].sample(min(1000, num_reviews), replace=True).tolist())

    # Usar el mismo vectorizador TF-IDF (o uno nuevo si quieres) para analizar el texto del cluster
    # Para simple frecuencia, puedes usar CountVectorizer o analizar la cadena directamente
    from collections import Counter
    import re

    words = re.findall(r'\b\w+\b', sample_text.lower()) # Extrae palabras, convierte a min√∫sculas
    stop_words = set(['el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'de', 'en', 'y', 'a', 'que', 'es', 'con', 'para', 'del', 'al', 'se', 'por', 'su', 'sus', 'como', 'mas', 'pero', 'este', 'esta', 'si', 'no', 'cuando', 'donde', 'porque', 'muy', 'mi', 'mis', 'tu', 'tus', 'el', 'ella', 'nosotros', 'vosotros', 'ellos', 'las', 'los']) # Ejemplo de stop words en espa√±ol

    # Filtrar stop words
    filtered_words = [word for word in words if word not in stop_words and len(word) > 1]

    word_counts = Counter(filtered_words)
    print("\nPalabras m√°s frecuentes (Top 10):")
    print(word_counts.most_common(10))

    print("-" * (len(f"--- Cluster {cluster_id} ---"))) # Separador para el siguiente cluster
```

Al analizar la matriz de confusi√≥n del modelo de clasificaci√≥n supervisado, observamos los siguientes tipos de errores:

1.  **Falsos Positivos (FP):** Rese√±as que el modelo predijo como "Positivo" (rating >= 4) pero que en realidad eran "Negativo" (rating < 4). Estos son comentarios que *parecen* positivos al modelo basado en las palabras clave, pero la intenci√≥n real del cliente era negativa.
2.  **Falsos Negativos (FN):** Rese√±as que el modelo predijo como "Negativo" pero que en realidad eran "Positivo". Estos son comentarios que *parecen* negativos al modelo, pero la intenci√≥n real del cliente era positiva (a pesar de quiz√°s usar un lenguaje cr√≠tico o mencionar un problema menor en una rese√±a mayoritariamente positiva).

Para entender c√≥mo estos errores se relacionan con los clusters, podemos analizar la distribuci√≥n de los FP y FN dentro de cada cluster identificado por K-means. Es probable que algunos clusters contengan una proporci√≥n m√°s alta de FP o FN que otros.

Por ejemplo:
*   **Cluster con alta proporci√≥n de FP:** Podr√≠a contener rese√±as que usan un lenguaje generalmente positivo pero que detallan un problema espec√≠fico grave que llev√≥ a un rating bajo. O rese√±as sarc√°sticas que el modelo no logr√≥ detectar.
*   **Cluster con alta proporci√≥n de FN:** Podr√≠a contener rese√±as que mencionan un defecto o problema al principio, lo que influye negativamente en la clasificaci√≥n TF-IDF, pero que luego concluyen con una evaluaci√≥n muy positiva del producto o servicio. O rese√±as muy cortas donde el contexto positivo no es suficiente para compensar palabras potencialmente negativas.
*   **Clusters bien clasificados:** Probablemente contienen rese√±as con lenguaje claro y directo que se alinea fuertemente con su rating (muy positivos o muy negativos).

Al perfilar los clusters, buscamos caracter√≠sticas distintivas (palabras clave, longitud, rating promedio, votos √∫tiles) que puedan explicar por qu√© las rese√±as en ciertos clusters son m√°s propensas a ser mal clasificadas.

**Relaci√≥n Espec√≠fica (Basada en el perfilado de clusters):**
